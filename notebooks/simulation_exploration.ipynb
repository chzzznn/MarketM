```python
import sys
sys.path.append('../src')  # So we can import from /src

from asset_classes import Card, CardType
from market_simulation import run_simulation_round

# Step 1: Define your assets
assets = [
    Card("TechCorp", CardType.STOCK, 100, 5),
    Card("Gold Futures", CardType.COMMODITY, 80, 3),
    Card("GovBond 10Y", CardType.BOND, 90, 2)
]

# Step 2: Simulate a market round
print("Market Starting:")
for a in assets:
    print(a)

run_simulation_round(assets)
```

```python
import matplotlib.pyplot as plt

prices = {asset.name: [asset.current_price] for asset in assets}
# 10 rounds run test
for _ in range(10): 
    run_simulation_round(assets)
    for a in assets:
        prices[a.name].append(a.current_price)

# Plotting
for name, price_history in prices.items():
    plt.plot(price_history, label=name)
plt.legend()
plt.title("Asset Prices Over Time")
plt.xlabel("Round")
plt.ylabel("Price")
plt.show()
```
